Esto se utiliza para modificar algunos archivos yaml, ya que la libreria que resuelve el data.yaml a veces no lo puede resolver correctamente

```python
import yaml
# Por alguna razon, el config esta mal armado, lo tengo que corregir
def modify_yaml_property(file_path, property_path, new_value):
    # Load YAML file
    with open(file_path, 'r') as yaml_file:
        data = yaml.safe_load(yaml_file)

    # Traverse the property path
    keys = property_path.split('.')
    current_data = data
    for key in keys[:-1]:
        if key in current_data:
            current_data = current_data[key]
        else:
            raise KeyError(f"Property path '{property_path}' is not valid.")

    # Modify the specified property
    current_data[keys[-1]] = new_value

    # Save the modified data back to the YAML file
    with open(file_path, 'w') as yaml_file:
        yaml.dump(data, yaml_file, default_flow_style=False)

# Example usage
yaml_file_path = 'your_yaml_file.yaml'
property_to_modify = 'nested.property.test'
new_value_for_property = 'new_value_for_test'

modify_yaml_property("./data.yaml", "test", "./test/images")
modify_yaml_property("./data.yaml", "train", "./train/images")
modify_yaml_property("./data.yaml", "val", "./valid/images")
```
